```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python

def random_walk(steps):
    # Initialisiere den Random Walk mit einer Startposition von 0
    walk = [0.5]

    # Führe den Random Walk durch
    for _ in range(steps):
        # Generiere einen normalverteilten Schritt zwischen 0 und 1
        step = np.random.normal(loc=0, scale=0.1)  # Mittelwert (loc) von 0 und Standardabweichung (scale) von 0.15

        # Füge den Schritt zur aktuellen Position hinzu
        position = walk[-1] + step
        # Begrenze die Position auf das Intervall [0, 1]
        position = max(0.1, min(0.9, position))
        # Füge die Position zum Random Walk hinzu
        walk.append(position)

    return walk

data = pd.read_csv("C:/Users\sarah/two-armed-bandit/two-armed-bandit/orders/actual_experiment.csv", sep=',')
data = pd.DataFrame(data)
#print(data)
#print(data.loc[:,"stim_0_value"])
#print(random_walk(len(data.loc[:,"stim_0_value"])).type)
probs_stim_0 = random_walk(len(data.loc[:,"blue_stim_value"])-1)
data.loc[:,"blue_stim_value"] = pd.Series(probs_stim_0, dtype='float64')
probs_stim_1 = random_walk(len(data.loc[:,"red_stim_value"])-1)
data.loc[:,"red_stim_value"] = pd.Series(probs_stim_1, dtype='float64')
data.to_csv("C:/Users\sarah/two-armed-bandit/two-armed-bandit/orders/actual_experiment.csv",index=False)
#print(data.loc[:,"stim_1_value"])

plt.plot(probs_stim_0, color='blue', label='_Blue stimulus')
plt.plot(probs_stim_1, color='red', label='_Red Stimulus')
plt.xlabel('Steps')
plt.ylabel('Value')
plt.title('Probabilites for winning')
plt.legend()
plt.grid(True)
plt.show()
```
